```python
import emd
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sampleRate = 200 # Hz
totalTime = 5 # seconds
t = np.linspace(0, totalTime, sampleRate*totalTime)
x = 0.7*np.sin(2*np.pi*5*t) + np.sin(2*np.pi*20*t)
```

```python
plt.figure(figsize=(16,6))
plt.scatter(t, x)
plt.plot(t, x)
plt.xlim(0, 1);
```

```python
imf = emd.sift.sift(x)
fig = plt.figure(figsize=(12,6))
emd.plotting.plot_imfs(imf, cmap=False, scale_y=True, fig=fig)
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf, sample_rate=sampleRate, method='hilbert')
```

```python
freq_edges, freq_centres = emd.spectra.define_hist_bins(0, 30, len(x)*2)
# Amplitude weighted HHT per IMF
f, spec_weighted = emd.spectra.hilberthuang(IF, IA, freq_edges, sum_imfs=False)

# Unweighted HHT per IMF - we replace the instantaneous amplitude values with ones
f, spec_unweighted = emd.spectra.hilberthuang(IF, np.ones_like(IA), freq_edges, sum_imfs=False)
```

```python
plt.figure(figsize=(16, 6))
plt.subplots_adjust(wspace=0.3)
plt.subplot(121)
plt.plot(freq_centres, spec_unweighted)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Count')

plt.title('unweighted\nHilbert-Huang Transform')

plt.subplot(122)
plt.plot(freq_centres, spec_weighted)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

plt.title('IA-weighted\nHilbert-Huang Transform')


plt.show()
```

```python
f, hht = emd.spectra.hilberthuang(IF, IA, freq_edges, mode='power', sum_time=False) # now sum_time =False, 1d sum_imfs = false

time_centers = (np.arange(len(x))-0.5)/sampleRate  # Is this the correct way of getting back to the 5 seconds?

#smoothing for visualization
shht = ndimage.gaussian_filter(hht, 0.2)

plt.figure(figsize=(8,6))
plt.pcolormesh(time_centers, freq_edges[0:-1], shht, cmap='nipy_spectral', vmax=1, vmin=0)
cb = plt.colorbar()
cb.set_label('Power', rotation=90)
```
